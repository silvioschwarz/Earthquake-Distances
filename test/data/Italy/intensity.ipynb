```python
Ix =[]
meann = []
ssd = []




for i in range(10,5,-1):
    Ix.append(i)
    meann.append( nf.loc[nf.Is==i].R.mean())
    ssd.append(nf.loc[nf.Is==i].R.std())
    print "Ix: %d mean %s: sd: %s" % (i, nf.loc[nf.Is==i].R.mean(), nf.loc[nf.Is==i].R.std())
```

```python
plt.plot(Ix, meann)

from scipy.optimize import curve_fit
def func(x, a,b,c):
        return a*np.exp(-b*x)+c

    
x = np.linspace(10,6,5)
    
popt, pcov = curve_fit(func, x, meann)
plt.plot(x, func(x, *popt), 'r',
          label='fit: a=%5.3f b=%5.3f c=%5.3f' % tuple(popt))
popt
```

```python
plt.plot( Ix,ssd)

x = np.linspace(10,6,5)
    
popt, pcov = curve_fit(func, x, ssd)
plt.plot(x, func(x, *popt), 'r',
          label='fit: a=%5.3f b=%5.3f c=%5.3f' % tuple(popt))
popt
```
